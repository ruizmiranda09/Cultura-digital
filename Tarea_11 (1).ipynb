## Parte 3: Condicionales en Python

### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
var1 = int(input("Dame una hora"))
if(var1<12):
  print('Es mañana')
else:
   print("Es tarde o noche")
```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python
var1 = int(input("Dame un numero "))
var2 = int(input("Dame otro numero "))
if(var1>var2):
  print(var1, "es mayor que", var2)
elif(var1==var2):
  print(var1, "es igual a ", var2)
else:
  print(var1, "es menor que ", var2)
```

### Ejercicio 3
